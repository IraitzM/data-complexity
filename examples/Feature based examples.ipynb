```python
import numpy as np

n_samples = 300
n = 1 # trials
p = 0.5
y = np.random.binomial(n, p, n_samples)

x1 = np.random.random(n_samples)
x2 = 0.5+(-1+y)*0.20 # x2 sets 0 samples on 0.20 value and 1 on 0.70

X = np.concatenate((x1[:, None], x2[:, None]), axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x=X[:, 0], y=X[:, 1], alpha=0.6, hue=y, palette="deep")
plt.show()
```

```python
from dcm.feature_based import FeatureBasedMeasures

feat = FeatureBasedMeasures()
feat.fit(X, y)
```

```python
feat.transform()
```
